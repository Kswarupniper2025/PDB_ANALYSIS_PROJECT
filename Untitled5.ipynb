```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
class SimplePNN:
    def __init__(self, sigma=0.5):
        self.sigma = sigma

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        self.classes = np.unique(y)
        return self

    def _prob(self, x, X_class):
        diff = X_class - x
        return np.mean(np.exp(-np.sum(diff**2, axis=1) / (2 * self.sigma**2)))

    def predict(self, X):
        y_pred = []
        for x in X:
            probs = []
            for c in self.classes:
                X_class = self.X_train[self.y_train == c]
                probs.append(self._prob(x, X_class))
            y_pred.append(np.argmax(probs))
        return np.array(y_pred)
```

```python
model = SimplePNN(sigma=0.5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

```python
df = pd.DataFrame({'True': y_test, 'Predicted': y_pred})
print(df.head())
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='viridis', edgecolor='k')
plt.title(f"PNN Predictions (Accuracy: {acc:.2f})")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
for s in [0.1, 0.3, 0.5, 1, 2,3,4,5,6,7]:
    model = SimplePNN(sigma=s)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("Sigma =", s, "Accuracy =", accuracy_score(y_test, y_pred))
```
